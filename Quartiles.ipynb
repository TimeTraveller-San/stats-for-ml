# Quartiles and IQR
Quartiles, as the name suggests divide the sample into 4 parts.

- Sort the data
- Find median (Mark this as quartile Q2)
- Find median of left sample (Mark this as quartile Q1)
- Find median of right sample (Mark this as quartile Q3)

An important measure associated with this is the IQR (inter quartile range) which is given by :
> IQR = Q3 - Q1

IQR gives a general idea of the range and localization of our sample. Better than these banal numbers, we plot interesting graphs called - box and whisker plot

# Python 

## Imports

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
sns.set_style('dark')
```

## Load dataset

```python
iris = datasets.load_iris()
X = pd.DataFrame(iris.data)
y = iris.target
```

```python
for i, col in enumerate(X.columns):
    q1, q2, q3 = X[col].quantile(0.25),\
                              X[col].quantile(0.5),\
                              X[col].quantile(0.75)
    IQR = q3 - q1
    print(f"Feature {col} \n Q1: {q1} \t Q2: {q2} \t Q3: {q3} \t IQR: {IQR}")
```

### Box plot

```python
for i, col in enumerate(X.columns):
    plt.figure(i)
    sns.boxplot(X[col], palette="inferno")
    plt.title(f"Feature: {col}")
```

## Box plot with histograms

```python
for i, col in enumerate(X.columns):
    plt.figure(i)
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (1, 4)})
    plt.title(f"Feature: {col}")
    sns.boxplot(X[col], palette="coolwarm", ax=ax_box)
    ax_box.set(xlabel='')
    sns.distplot(X[col], ax=ax_hist)
    ax_hist.set(xlabel='')
```
